```python
import pandas as pd
import numpy as np
data_path='/Users/kennetharguedas/Desktop/mos_jfk.csv'
cols = ['station', 'model', 'runtime', 'ftime', 'n_x', 'tmp', 'dpt', 'cld',
       'wdr', 'wsp', 'p06', 'p12', 'q06', 'q12', 't06', 't12', 'snw', 'cig',
       'vis', 'obv', 'poz', 'pos', 'typ']
df=pd.read_csv(data_path, names=cols)
```

```python
df.head()
```

```python
cdf = df.drop(0, axis=0)
```

```python
cdf
```

```python
cdf['date'] = pd.to_datetime(cdf['runtime'])
```

```python
cdf['tmp'] = cdf['tmp'].astype(float)
```

```python
cdf.info()
```

```python
cdf['date'].dt.time.head()
```

```python
hourly = cdf[['date','tmp']].groupby(cdf['date'].dt.time)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(15,5))

for hr, dfh in hourly:
    dfh.set_index('date').plot(ax=ax, label=hr)
_ = ax.legend(['0', '6', '12', '18'])
```
